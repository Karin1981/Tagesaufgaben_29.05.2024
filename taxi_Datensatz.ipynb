```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
taxis = sns.load_dataset("taxis")
taxis.to_csv("taxis.csv")
taxis.head()
```

```python
taxis.boxplot(column="total")

plt.title("Boxplot der Gesamtkosten")
plt.xlabel("Gesamtkosten")


plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=taxis["total"])

plt.title("Gesamtkosten")
plt.xlabel("Gesamtkosten")
plt.show()
```
